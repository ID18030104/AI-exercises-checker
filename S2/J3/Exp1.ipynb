```python
import pandas as pd
from sklearn.impute import SimpleImputer



# ðŸ” 1. Afficher les colonnes avec valeurs manquantes
print("Valeurs manquantes par colonne :")
print(titanic_df.isnull().sum())

# ðŸ§¹ 2. Supprimer les lignes avec valeurs manquantes (exemple pour vÃ©rification, mais on ne va pas garder Ã§a)
titanic_dropna = titanic_df.dropna()
print("\nTaille aprÃ¨s suppression des lignes avec valeurs manquantes :", titanic_dropna.shape)

# ðŸ§  3. Imputation : remplir la colonne "Age" avec la moyenne
age_imputer = SimpleImputer(strategy='mean')
titanic_df['Age'] = age_imputer.fit_transform(titanic_df[['Age']])

# ðŸ§± 4. Remplir la colonne "Cabin" avec une valeur constante (ex : 'U' pour Unknown)
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('U')

# ðŸ§­ 5. Remplir la colonne "Embarked" avec le mode (la valeur la plus frÃ©quente)
embarked_imputer = SimpleImputer(strategy='most_frequent')
titanic_df['Embarked'] = embarked_imputer.fit_transform(titanic_df[['Embarked']]).ravel()

# âœ… VÃ©rification finale
print("\nValeurs manquantes aprÃ¨s tous les traitements :")
print(titanic_df.isnull().sum())
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder

# 1. CrÃ©er la colonne FamilySize
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1

# 2. Extraire le Title depuis la colonne Name
titanic_df['Title'] = titanic_df['Name'].apply(lambda name: name.split(',')[1].split('.')[0].strip())

# 3. Encodage des colonnes catÃ©gorielles avec LabelEncoder
le_sex = LabelEncoder()
le_embarked = LabelEncoder()
le_title = LabelEncoder()

titanic_df['Sex_encoded'] = le_sex.fit_transform(titanic_df['Sex'])
titanic_df['Embarked_encoded'] = le_embarked.fit_transform(titanic_df['Embarked'])
titanic_df['Title_encoded'] = le_title.fit_transform(titanic_df['Title'])

# 4. Normalisation de la colonne Age
scaler = MinMaxScaler()
titanic_df['Age_normalized'] = scaler.fit_transform(titanic_df[['Age']])

# 5. Afficher les premiÃ¨res lignes pour vÃ©rification
print(titanic_df[['FamilySize', 'Title', 'Sex_encoded', 'Embarked_encoded', 'Title_encoded', 'Age', 'Age_normalized']].head())
```

```python
# DÃ©tection et suppression des outliers avec l'IQR

# 1. Pour la colonne Fare
Q1_fare = titanic_df['Fare'].quantile(0.25)
Q3_fare = titanic_df['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
lower_fare = Q1_fare - 1.5 * IQR_fare
upper_fare = Q3_fare + 1.5 * IQR_fare

# 2. Pour la colonne Age
Q1_age = titanic_df['Age'].quantile(0.25)
Q3_age = titanic_df['Age'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_age = Q1_age - 1.5 * IQR_age
upper_age = Q3_age + 1.5 * IQR_age

# 3. Garder uniquement les lignes sans outliers
titanic_df_no_outliers = titanic_df[
    (titanic_df['Fare'] >= lower_fare) & (titanic_df['Fare'] <= upper_fare) &
    (titanic_df['Age'] >= lower_age) & (titanic_df['Age'] <= upper_age)
]

# 4. Affichage des tailles avant/aprÃ¨s
print("Taille originale :", titanic_df.shape)
print("Taille sans outliers :", titanic_df_no_outliers.shape)
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# On travaille sur une copie
df_scaled = titanic_df.copy()

# 1. Standardisation de la colonne Fare
scaler_standard = StandardScaler()
df_scaled['Fare_standardized'] = scaler_standard.fit_transform(df_scaled[['Fare']])

# 2. Normalisation de la colonne Age
scaler_minmax = MinMaxScaler()
df_scaled['Age_normalized'] = scaler_minmax.fit_transform(df_scaled[['Age']])

# 3. Affichage des premiÃ¨res lignes
print(df_scaled[['Fare', 'Fare_standardized', 'Age', 'Age_normalized']].head())
```

```python
import pandas as pd

# Copie du DataFrame
df_age_bins = titanic_df.copy()

# ðŸ§± 1. CrÃ©er les groupes d'Ã¢ge avec pd.cut()
age_bins = [0, 12, 18, 35, 60, 100]
age_labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Senior']
df_age_bins['AgeGroup'] = pd.cut(df_age_bins['Age'], bins=age_bins, labels=age_labels)

# ðŸŽ¯ 2. Appliquer le one-hot encoding sur les groupes d'Ã¢ge
df_age_encoded = pd.get_dummies(df_age_bins, columns=['AgeGroup'])

# âœ… Afficher un aperÃ§u
print(df_age_encoded[['Age'] + [col for col in df_age_encoded.columns if "AgeGroup" in col]].head())
```
