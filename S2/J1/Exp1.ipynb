```python
import pandas as pd

print("pandas version:", pd.__version__)
```

```python
# Create the bookstore sales dataset
data = {
    'Book Title': ['The Great Gatsby', 'To Kill a Mockingbird', '1984', 'Pride and Prejudice', 'The Catcher in the Rye'],
    'Author': ['F. Scott Fitzgerald', 'Harper Lee', 'George Orwell', 'Jane Austen', 'J.D. Salinger'],
    'Genre': ['Classic', 'Classic', 'Dystopian', 'Classic', 'Classic'],
    'Price': [10.99, 8.99, 7.99, 11.99, 9.99],
    'Copies Sold': [500, 600, 800, 300, 450]
}

df = pd.DataFrame(data)

# Explore the dataset
print("First few rows:")
print(df.head())

print("\nStatistical summary:")
print(df.describe())

print("\nInfo summary:")
print(df.info())

# Sort by Price
print("\nSorted by Price:")
print(df.sort_values(by='Price'))

# Sort by Copies Sold
print("\nSorted by Copies Sold:")
print(df.sort_values(by='Copies Sold', ascending=False))

# Filter by Genre 'Classic'
print("\nBooks in Classic genre:")
print(df[df['Genre'] == 'Classic'])

# Filter books with Price above 10
print("\nBooks with Price above 10:")
print(df[df['Price'] > 10])

# Group by Author and sum Copies Sold
print("\nCopies Sold by Author:")
print(df.groupby('Author')['Copies Sold'].sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Dataset\Iris Dataset\Iris Dataset\Iris.csv")
```

```python
mean = df['SepalLengthCm'].mean()
median = df['SepalLengthCm'].median()
mode = df['SepalLengthCm'].mode()[0]  
```

```python
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['SepalLengthCm'], kde=True, bins=10, color="skyblue")
plt.title('Distribution of Sepal Length (cm)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```
