# 📘 Exercise 1: Introduction to Data Analysis

## 🧠 Qu’est-ce que l’analyse de données ?

L’analyse de données est le processus qui consiste à **collecter, organiser, nettoyer, explorer et interpréter** des données dans le but d’en tirer des informations utiles. Elle peut être descriptive (comprendre ce qui s’est passé), diagnostique (pourquoi cela s’est produit), prédictive (ce qui va probablement arriver) ou prescriptive (ce qu’il faut faire). Elle s’appuie sur des techniques statistiques, des algorithmes, et des outils visuels pour aider à la prise de décision.

---

## 🌍 Pourquoi l’analyse de données est-elle importante aujourd’hui ?

À l’ère du numérique, des milliards de données sont générées chaque jour (réseaux sociaux, capteurs, transactions, navigation web, etc.). **L’analyse de données permet de transformer ce flot d’informations brutes en connaissance exploitable**. Elle aide à :
- prendre des décisions éclairées et plus rapides,
- optimiser des processus complexes,
- détecter des tendances ou anomalies,
- améliorer la satisfaction client,
- et innover de manière ciblée.

Sans analyse, les données restent inutiles. Avec elle, elles deviennent un avantage stratégique.

---

## 🧪 Trois domaines d’application de l’analyse de données

### 1. 🏥 Santé publique
L’analyse des données hospitalières permet de suivre la propagation des maladies, d’optimiser les ressources médicales ou encore de personnaliser les traitements. Par exemple, pendant la pandémie de COVID-19, les données de contamination, d’hospitalisation et de vaccination ont guidé les politiques sanitaires.

### 2. 💼 Marketing & E-commerce
Les entreprises analysent les données de navigation, d’achat et de comportement pour **cibler les publicités**, **personnaliser l’expérience utilisateur** et **augmenter les ventes**. Amazon, par exemple, recommande des produits basés sur des algorithmes d’analyse comportementale.

### 3. 🏙️ Transports et mobilité
Des acteurs comme la SNCF ou Google Maps utilisent des données en temps réel pour **prévoir les retards**, **optimiser les itinéraires** ou encore **anticiper la maintenance** des véhicules, ce qui améliore la qualité du service pour les usagers.

---

## ✅ Conclusion

L’analyse de données est aujourd’hui **indispensable dans tous les secteurs**. Elle permet non seulement de mieux comprendre le monde, mais aussi de le rendre plus efficace, plus sûr et plus adapté aux besoins des individus. À l’heure où les données sont partout, savoir les analyser devient une compétence clé pour les citoyens et les professionnels de demain.

```python
# Create the bookstore sales dataset
data = {
    'Book Title': ['The Great Gatsby', 'To Kill a Mockingbird', '1984', 'Pride and Prejudice', 'The Catcher in the Rye'],
    'Author': ['F. Scott Fitzgerald', 'Harper Lee', 'George Orwell', 'Jane Austen', 'J.D. Salinger'],
    'Genre': ['Classic', 'Classic', 'Dystopian', 'Classic', 'Classic'],
    'Price': [10.99, 8.99, 7.99, 11.99, 9.99],
    'Copies Sold': [500, 600, 800, 300, 450]
}

df = pd.DataFrame(data)

# Explore the dataset
print("First few rows:")
print(df.head())

print("\nStatistical summary:")
print(df.describe())

print("\nInfo summary:")
print(df.info())

# Sort by Price
print("\nSorted by Price:")
print(df.sort_values(by='Price'))

# Sort by Copies Sold
print("\nSorted by Copies Sold:")
print(df.sort_values(by='Copies Sold', ascending=False))

# Filter by Genre 'Classic'
print("\nBooks in Classic genre:")
print(df[df['Genre'] == 'Classic'])

# Filter books with Price above 10
print("\nBooks with Price above 10:")
print(df[df['Price'] > 10])

# Group by Author and sum Copies Sold
print("\nCopies Sold by Author:")
print(df.groupby('Author')['Copies Sold'].sum())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Dataset\Iris Dataset\Iris Dataset\Iris.csv")
```

```python
mean = df['SepalLengthCm'].mean()
median = df['SepalLengthCm'].median()
mode = df['SepalLengthCm'].mode()[0]  
```

```python
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['SepalLengthCm'], kde=True, bins=10, color="skyblue")
plt.title('Distribution of Sepal Length (cm)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
```

##  Basic Data Analysis - Iris Dataset

###  Quantitative Column Analyzed: `SepalLengthCm`

**Statistical Measures:**
- **Mean**: 5.84 cm
- **Median**: 5.80 cm
- **Mode**: 5.0 cm

Ces trois mesures montrent que la distribution est légèrement asymétrique, avec quelques valeurs plus élevées qui tirent la moyenne au-dessus de la médiane.

## 🎯 Observation & Data Exploration – Sleep Dataset

### 🔍 Objectif :
Identifier des colonnes intéressantes pour des analyses spécifiques.

### 📁 Dataset : *How Much Sleep Do Americans Really Get?*

### Colonnes disponibles :
- `Year` : l’année de l’observation → utile pour une **analyse des tendances** (évolution du sommeil dans le temps).
- `Avg hrs per day sleeping` : moyenne d’heures de sommeil → variable principale à observer et à **comparer entre groupes**.
- `Type of Days` : différencie les jours (ex : "All days", "Weekdays", "Weekend") → pertinent pour une **comparaison des habitudes selon le type de jour**.
- `Sex` : genre → permet une **comparaison homme/femme**.
- `Age Group` : tranche d'âge → utile pour une **comparaison selon l’âge**.
- `Standard Error` : mesure de la variabilité → important pour évaluer la **fiabilité des moyennes**.

### 🧠 Choix d’analyse possibles :
1. **Trend Analysis (Analyse temporelle)** :
   - Colonnes utilisées : `Year` + `Avg hrs per day sleeping`
   - Objectif : voir si les Américains dorment plus ou moins avec le temps.

2. **Group Comparison (Comparaison de groupes)** :
   - Comparer `Avg hrs per day sleeping` en fonction de :
     - `Sex`
     - `Age Group`
     - `Type of Days`

3. **Fiabilité des mesures** :
   - En analysant `Standard Error`, on peut juger si les données sont stables ou très dispersées.

---

### ✅ Conclusion :
Pour des observations riches, on combinera `Avg hrs per day sleeping` avec les colonnes `Year`, `Sex`, `Age Group`, et `Type of Days` selon le type d’analyse souhaité.
