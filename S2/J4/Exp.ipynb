Exercice 2

```python
import matplotlib.pyplot as plt

temperatures = [72, 74, 76, 80, 82, 78, 75]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

plt.plot(days, temperatures, marker='o', linestyle='-', color='b')

plt.title("Weekly Temperature")
plt.xlabel("Days of the Week")
plt.ylabel("Temperature (°F)")

plt.show()
```

Exercice 4

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
df = pd.DataFrame({
    "CGPA": np.round(np.random.uniform(2.0, 4.0, 100), 2)
})

sns.histplot(data=df, x="CGPA", bins=10, color="#651FFF")  
plt.title("Distribution of CGPA among Students")
plt.xlabel("CGPA")
plt.ylabel("Number of Students")
plt.show()
```

Exercice 5

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {"Do you have Anxiety?": ["Yes", "No", "Yes", "Yes", "No", "Yes", "No", "Yes", "No", "Yes"],
    "Choose your gender": ["Male", "Female", "Female", "Male", "Female", "Female", "Male", "Female", "Male", "Male"]}

df = pd.DataFrame(data)

sns.countplot(data=df, x="Choose your gender", hue="Do you have Anxiety?", palette="pastel")
plt.title("Anxiety Levels by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.show()
```

Exercice 6

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = { "Age": [18, 19, 20, 20, 21, 22, 23, 24, 25, 26, 27],
    "Do you have Panic Attacks?": ["No", "No", "No","Yes", "Yes", "No", "Yes", "No", "Yes", "No", "Yes"]}

df = pd.DataFrame(data)

df["PanicBinary"] = df["Do you have Panic Attacks?"].map({"Yes": 1, "No": 0})

sns.scatterplot(data=df, x="Age", y="PanicBinary", s=100, color="#FF1744")
plt.title("Panic Attacks by Age")
plt.xlabel("Age")
plt.ylabel("Panic Attacks (Yes=1, No=0)")
plt.yticks([0, 1], ["No", "Yes"])  
plt.show()
```
