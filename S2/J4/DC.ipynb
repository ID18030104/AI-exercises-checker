```python
import pandas as pd


df_2019 = pd.read_csv("archive/2019.csv")


print(df_2019.head())

print(df_2019.info())
print(df_2019.columns)
```

```python

print(df_2019.isnull().sum())
df_2019_clean = df_2019.dropna()
print(df_2019_clean.isnull().sum())
```

```python
print(df_2019_clean.columns)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_2019_clean['Social support'], df_2019_clean['Score'], alpha=0.6)
plt.title("Social Support vs Happiness Score (2019)")
plt.xlabel("Social Support")
plt.ylabel("Happiness Score")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# On trie les pays par PIB pour plus de lisibilit
df_sorted = df_2019_clean.sort_values("GDP per capita")

fig, ax1 = plt.subplots(figsize=(18, 6))

# Bar plot pour le PIB par habitant (GDP per capita)
ax1.bar(df_sorted["Country or region"], df_sorted["GDP per capita"], color='skyblue', label='GDP per capita')
ax1.set_xlabel('Country or region')
ax1.set_ylabel('GDP per capita')
ax1.tick_params(axis='x', rotation=90)

# Courbe pour lesprance de vie en bonne sant (Healthy life expectancy)
ax2 = ax1.twinx()
ax2.plot(df_sorted["Country or region"], df_sorted["Healthy life expectancy"], color='green', marker='o', linewidth=2, label='Healthy life expectancy')
ax2.set_ylabel('Healthy life expectancy')

plt.title("GDP per Capita and Healthy Life Expectancy by Country (2019)")
plt.tight_layout()
plt.show()
```
