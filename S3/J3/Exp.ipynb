```python
from scipy import stats
import numpy as np


data = [12, 15, 13, 12, 18, 20, 22, 21]


mean_value = np.mean(data)
print("Moyenne :", mean_value)


median_value = np.median(data)
print("Médiane :", median_value)


variance_value = np.var(data)
print("Variance :", variance_value)


std_deviation = np.std(data)
print("Écart-type :", std_deviation)
```

Ex 3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

data = norm.rvs(loc=50, scale=10, size=1000)


plt.figure(figsize=(10, 5))

plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Histogramme des données')


x = np.linspace(min(data), max(data), 1000)


pdf = norm.pdf(x, loc=50, scale=10)

plt.plot(x, pdf, 'r-', label='Courbe normale théorique')


plt.title('Distribution normale (μ=50, σ=10)')
plt.xlabel('Valeurs')
plt.ylabel('Densité de probabilité')
plt.legend()


plt.show()
```

Ex 4

```python
import numpy as np
from scipy.stats import ttest_ind


data1 = np.random.normal(50, 10, 100)  
data2 = np.random.normal(60, 10, 100)  

t_stat, p_value = ttest_ind(data1, data2)


print(f"T-statistic : {t_stat:.2f}")
print(f"P-value     : {p_value:.4f}")


if p_value < 0.05:
    print("→ Résultat significatif : les deux échantillons ont des moyennes différentes.")
else:
    print("→ Pas de différence significative entre les deux échantillons.")
```

Ex 5

```python
import numpy as np
from scipy import stats


house_sizes = np.array([50, 70, 80, 100, 120])
house_prices = np.array([150000, 200000, 210000, 250000, 280000])


slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)


print(f"Pente (slope)     : {slope:.2f} € par m²")
print(f"Ordonnée (intercept) : {intercept:.2f} €")
print(f"R² (coefficient de corrélation au carré) : {r_value**2:.4f}")


size = 90
predicted_price = slope * size + intercept
print(f"Prix estimé pour une maison de {size} m² : {predicted_price:.2f} €")
```

Ex 6

```python
from scipy.stats import f_oneway


fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]


f_stat, p_value = f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)

print(f"F-value  : {f_stat:.4f}")
print(f"P-value  : {p_value:.4f}")


if p_value < 0.05:
    print("✅ Il existe une différence significative entre les engrais.")
else:
    print("❌ Aucune différence significative détectée entre les engrais.")
```

Ex 8

```python
import pandas as pd
from scipy.stats import pearsonr, spearmanr

data = pd.DataFrame({
    'age': [23, 25, 30, 35, 40],
    'income': [35000, 40000, 50000, 60000, 70000]
})


pearson_corr, pearson_p = pearsonr(data['age'], data['income'])


spearman_corr, spearman_p = spearmanr(data['age'], data['income'])


print(f"Corrélation de Pearson : {pearson_corr:.4f} (p = {pearson_p:.4f})")
print(f"Corrélation de Spearman : {spearman_corr:.4f} (p = {spearman_p:.4f})")
```
